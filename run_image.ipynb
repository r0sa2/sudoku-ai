Notebook to demo how to use the system for images

```python
import cv2
import matplotlib.pyplot as plt

from ar_sudoku_solver.ar_sudoku_solver import ARSudokuSolver
```

Load and display image

```python
image = cv2.imread("./sudoku_dataset/image1062.jpg")
plt.imshow(cv2.cvtColor(image, code=cv2.COLOR_BGR2RGB))
plt.axis('off')
```

Solve and display solved image

```python
solved_image = ARSudokuSolver().solve_image(image)
plt.imshow(cv2.cvtColor(solved_image, code=cv2.COLOR_BGR2RGB))
plt.axis('off')
```
